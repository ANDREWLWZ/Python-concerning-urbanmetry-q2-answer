```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'C:\Users\User\Downloads/urban.xlsx')
```

```python
df.head(100)
```

```python
# Group the data by building and year, and calculate the median price psf for each group
```

```python
median_vprice_psf = df.groupby(["NAME", "YEAR"])["Valuation price"].median()
```

```python

# Plot the median price psf by building and year
```

```python
median_vprice_psf.unstack().plot(kind='line', stacked=False)
plt.title("Median Vluation Price PSF Across Time")
plt.xlabel("Year")
plt.ylabel("Median Price PSF")
```

```python
# Group the data by building and year, and calculate the average price psf for each group
```

```python
average_price_psf = df.groupby(["NAME", "YEAR"])["Built up price psf"].mean()
```

```python
average_price_psf.unstack().plot(kind='line', stacked=False)
plt.title("Average Price PSF Across Time")
plt.xlabel("Year")
plt.ylabel("Average Price PSF")
```
